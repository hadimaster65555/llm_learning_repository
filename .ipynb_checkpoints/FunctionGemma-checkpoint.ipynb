# Create Agent using FunctionGemma

Here I build simple agent that call goweather.xyz API to measure temperature given city name by leveraging LangChain and Ollama.

# Load Library

```python
import os
from dotenv import load_dotenv
import requests

from langchain_ollama import ChatOllama
from langchain_core.tools import tool
from langchain.agents import create_agent
```

# Define Model

```python
# define model
llm = ChatOllama(
    model="functiongemma:270m-it-fp16",
    validate_model_on_init=True,
    temperature=1,
    num_predict=256,
)
```

# Get Weather Agent

## Define Tools

```python
@tool
def get_weather(query: str) -> str:
    """
    Use this tool to get current weather for a city
    """
    url = f"https://goweather.xyz/v2/weather/{query}"
    response = requests.get(url)
    temperature = response.json().get("temperature")
    if temperature:
        return f"Temperature : {temperature}"
    return "No data found."

tools = [get_weather]
```

## Define Agent

```python
agent = create_agent(
    llm,
    tools=tools,
    system_prompt="You are a helpful AI assistant. Use tools when they are useful, especially for factual questions"
)
```

## Run Agent and Check Result

```python
result = agent.invoke(
    {"messages": [{"role": "user", "content": "What's the weather in Jakarta?"}]}
)
```

```python
result['messages']
```

```python
result['messages'][-1]
```

# Calculator Agent

## Define Tools

```python
@tool
def add(a: float, b: float) -> float:
    """Adds two numbers together."""
    return a + b

@tool
def subtract(a: float, b: float) -> float:
    """Subtracts b from a."""
    return a - b

@tool
def multiply(a: float, b: float) -> float:
    """Multiplies a with b."""
    return a * b

@tool
def divide(a: float, b: float) -> float:
    """Divides a by b."""
    if b == 0:
        return "Error: Division by zero."
    return a / b

# List of tools available to the agent
tools = [add, subtract, multiply, divide]
```

```python
agent = create_agent(
    llm,
    tools=tools,
    system_prompt="You are a helpful mathematical assistant. Use your tools to perform calculations."
)
```

```python
tools
```

```python
result = agent.invoke(
    {"messages": [{"role": "user", "content": "What is the result of 10000 add by 10?"}]}
)

result['messages'][-1]
```
